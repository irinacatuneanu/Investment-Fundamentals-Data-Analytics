```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_mean = 170
rev_stdev = 20
iterations = 1000
```

# Simulation of future revenues

```python
rev = np.random.normal(rev_mean, rev_stdev, iterations)
rev
```

# Plot the future revenues

```python
plt.figure(figsize = (15,6))
plt.plot(rev)
```

# Simulation of COGS

```python
COGS = - (rev * np.random.normal(0.6,0.1))
```

```python
plt.figure(figsize = (15,6))
plt.plot(COGS)
```

```python
COGS.mean()
```

```python
COGS.std()
```

# Gross Profit

```python
Gross_Profit = rev + COGS
Gross_Profit
```

```python
plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
```

```python
max(Gross_Profit)
```

```python
min(Gross_Profit)
```

```python
Gross_Profit.mean()
```

```python
Gross_Profit.std()
```
